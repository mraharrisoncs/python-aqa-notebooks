| Term or Symbol | Name              | Purpose                                                  |
|----------------|-------------------|----------------------------------------------------------|
| `+`            | Addition          | Add both operands                                        |
| `-`            | Subtraction       | Subtract second operand from first                       |
| `*`            | Multiplication    | Multiply two operands                                    |

| `%`            | Modulus           | Find the remainder after integer division               |
| `**`           | Exponent          | Raise to the power of                                   |
| Operator       | N/A               | A symbol for an operation such as `+`, `-`, `*`, `/`     |
| Operand        | N/A               | Terms either side of the operator, e.g. `3` and `4` in `3 + 4` |
| Integer        | N/A               | A whole number such as `7`, `-1`                         |
| Float          | N/A               | A number with a decimal part, also called Real, such as `4.0`, `3.14`, `9.999999` |

### Common misconceptions 1

##  

| ❌ Misconception                                | 🧪 What Students Might Do                     | ✅ Explanation / Fix                                                                                          |
| ---------------------------------------------- | --------------------------------------------- | ------------------------------------------------------------------------------------------------------------ |
| `/ 0`, `// 0`, and `% 0` are valid             | `average = total / countmodulus = number % 0` | Causes `ZeroDivisionError`. Always check that the denominator is not zero before using `/`, `//`, or `%`.    |
| `x` means multiply                             | `area = length x width`                       | Python treats `x` as a variable, not an operator. Use `*` for multiplication.                                |
| You can add or concatenate strings and numbers | `print("Score: " + score)`                    | You can't concatenate a string and an integer with `+`. Use `str(score)` or commas: `print("Score:", score)` |

### Common misconceptions 2

| ❌ Misconception                      | 🧪 What Students Might Do      | ✅ Explanation / Fix                                                                                              |
| ------------------------------------ | ------------------------------ | ---------------------------------------------------------------------------------------------------------------- |
| Operators work left to right         | `4 + 2 * 3`Expecting `18`      | Python follows precedence: `2 * 3 = 6`, then `4 + 6 = 10`. Use brackets to override precedence if needed.        |
| `%` gives a percentage               | `percent = score % total`      | `%` gives the remainder, not a percentage. Use: `percent = score / total * 100`                                  |
| Multiple `**` evaluate left to right | `2 ** 3 ** 2`Expecting `64`    | Python evaluates exponentiation right to left: `3 ** 2 = 9`, then `2 ** 9 = 512`. Use brackets to control order. |
| Mixed units                          | `total = hours * 60 + seconds` | Keep units consistent. Convert all values to the same unit (e.g. minutes or seconds) before calculating.         |
