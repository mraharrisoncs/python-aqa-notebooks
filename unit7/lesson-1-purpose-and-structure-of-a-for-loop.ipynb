### Why count-controlled iteration?

Count-controlled iteration, also known as **definite** iteration, repeats a block of code a specific number of times. This type of iteration is useful when the number of iterations is **known before the loop starts**. In Python, we use a `for` loop for definite iteration.

### When to use a for loop

We use a **for** loop when we know how many times we want to repeat a block of code. We might want to run a **set number of times** or process a **known list of items.** This is why we call it **definite** iteration, the loop will run a definite number of iterations.

> Remember, if we do **not know** how many times we want to loop, we would use **indefinite** iteration, also known as **condition-controlled** iteration. In Python this means a **while** loop which we met in Unit 5.

### Structure of a `for` Loop

The basic structure of a `for` loop in Python is as follows:

```python
for variable in sequence:
    # Code block to execute
```

* **`variable`**: This is a temporary variable that takes the value of the next item in the sequence on each iteration.

* **`sequence`**: This is the collection of items you want to iterate over. Usually this is a list **literal** or list **variable.**

* **Code block**: This is the block of code that will be executed on each iteration. It is indented to show that it belongs to the loop.

### Example for loop

Here is a simple example of a `for` loop iterating over a list:

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```


### Uses of a for loop

**Iterating over a range of numbers**: Performing an action a fixed number of times such as:

* displaying a times table or a countdown

* drawing a regular shape on the screen such as a square or hexagon

* constructing a string (e.g. a password) from a fixed number of random characters

**Processing items in a list**: Iterating over each item in a **list** helps perform operations like:

* calculating the sum or average of a list of numbers

* summing, searching or sorting a list of data 

* modifying every element in a list (such as adding 10% for inflation!)

### Definite versus indefinite

Here is a reminder of the key features of definite iteration (e.g. `for`) and  indefinite iteration (e.g. `while`):

| Definite iteration                                    | Indefinite iteration                                        |
| ----------------------------------------------------- | ----------------------------------------------------------- |
| Also known as **count-controlled** iteration          | Also known as **condition-controlled** iteration.           |
| Iterates over a list or a **range of numbers.**       | Loops until **a condition is met** or becomes unmet.        |
| Runs a **fixed** number of times                      | Number of iterations is **unknown** on entry to the loop    |
| Uses **`for`** keyword                                | Uses **`while`** keyword or **`repeat-until`** (not Python) |
| **Use cases of definite iteration**                   | **Use cases of indefinite iteration**                       |
| Performing an action a set number of times.           | Continuously checking for a condition to become true.       |
| Outputting a times table, countdown or regular shape. | Waiting for user input to match a specific condition.       |
| Sorting a list of known length.                       | Monitoring a sensor until a threshold is reached.           |
| Updating a list with a 20% reduction for each item.   | Repeatedly prompting for correct input until valid.         |

### A Simple Caesar Cipher

A **Caesar cipher** is one of the oldest and simplest ways to encode a message. It works by **shifting each letter** forward in the alphabet by a fixed number of places.

 

For example. with a shift of 3, all letters are moved forward 3 characters in the alphabet, so `a` becomes `d`, `b` becomes `e` and so on. If we go past `z`, the cipher **wraps around** to the beginning:

```
# Caesar cipher, shift = 3

a → d
b → e
c → f
...
x → a
y → b
z → c

```

Because a string is just a sequence of characters, a `for` loop is perfect for processing each one.

### Caesar Cipher Encoder Function

```python
def caesarEncode(message, shift):
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    result = ""

    for char in message:
        if char in alphabet:
            position = alphabet.index(char)
            newPosition = (position + shift) % 26
            result += alphabet[newPosition]
        else:
            result += char

    return result

# Example use
print(caesarEncode("hello world", 3))
```

### Explanation

* The function takes a string `message` and an integer `shift` as parameters.

* `alphabet` is assigned a simple string containing all lowercase letters.

* The loop processes the message **one character at a time**, storing the current character in `char.`

* If `char` is found in `alphabet`:

  * `alphabet.index(char)` finds its position in the alphabet, storing this in `position`.

  * `(position + shift) `    adds the shift to this position, calculating the position of the encoded character.

  * `% 26` applies **modulus 26**, to this new position, which  **wraps around** to the start of the alphabet. This has the same effect as this, but in less code:

    ```python
    if newPosition > 26: 
       newPosition = newPosition - 26
    ```

  * The new shifted letter is concatenated to the `result` string.

* If the character is **not** a letter (e.g. if it is a space or punctuation), it is copied unchanged.

* The function returns the encoded message.

### Example output



````python
# caesar_cipher.py
def caesarEncode(message, shift):
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    result = ""

    for char in message:
        if char in alphabet:
            position = alphabet.index(char)
            newPosition = (position + shift) % 26
            result += alphabet[newPosition]
        else:
            result += char

    return result

# Example use
print(caesarEncode("hello world", 3))
````

### Python practical: iterating over a list

Let's look at some examples of simple programs that iterate over lists, then you can write your own!

### Example 1: Sum of Numbers in a List

```python
def sumOfNumbers(numbers):
    total = 0
    for number in numbers:
        total += number
    return total

# Example use
numbers = [5, 10, 15, 20]
print("Sum of numbers:", sumOfNumbers(numbers))
```

### Explanation

* The function `sum_of_numbers` takes a list of numbers as its parameter.

* It initializes a variable `total` to 0, which will hold the sum of the numbers.

* The `for` loop iterates over each `number` in the `numbers` list.

* During each iteration, the current `number` is added to `total`.

* After the loop completes, the function returns the `total` sum of the numbers.

* The example use demonstrates calling the function with a list `[5, 10, 15, 20]` and printing the result.

***

### Example 2: Make Words Uppercase in a List

```python
def makeUppercase(words):
    upperWords = []
    for word in words:
        upperWords.append(word.upper())
    return upperWords

# Example use
words = ["python", "loop", "iteration", "code"]
print("Uppercase words:", makeUppercase(words))
```

### Explanation

* The function `make_uppercase` takes a list of words as its parameter.

* It initializes an empty list `upper_words` to store the uppercase versions of the words.

* The `for` loop iterates over each `word` in the `words` list.

* During each iteration, the `word.upper()` method is called to convert the word to uppercase, and the result is appended to `upper_words`.

* After the loop completes, the function returns the list `upper_words` containing all the uppercase words.

* The example use demonstrates calling the function with a list `["python", "loop", "iteration", "code"]` and printing the result.
