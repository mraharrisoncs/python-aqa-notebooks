### While loops for validation of input

Input validation is a crucial part of programming, ensuring input data is correct and meaningful. While loops are perfect for this task.

## Why validate input?

When users provide input, it may not always be in the format or range you expect. For example, if your program requires a number between 1 and 100, a user might accidentally enter a number outside this range, such as 0 or 1000. Input validation helps prevent errors and ensures your program runs smoothly.

## Using While Loops for Validation

A `while` loop can repeatedly prompt the user for input until they provide a valid response. Here's a basic example:

```python
# Example: Validating a test score between 1 and 100
def getScore():
  score = int(input("Enter a score between 1 and 100: "))
  while score < 1 or score > 100:
      print("Invalid input. Please try again.")
      score = int(input("Enter a score between 1 and 100: "))
  return score
```

Here, any numbers outside the range of 1 - 100 will cause the loop to iterate, and ask for another input. This continues until the user enters a number in the correct range. The while loop then terminates and the function returns this value.

## Boolean flag version

```python
# Example: Validating a test score between 1 and 100 using a Boolean flag
def getScore():
  erroneous = True
  while erroneous:
      score = int(input("Enter a score between 1 and 100: "))
      if score >= 0 and score <= 100:
          erroneous = False
  return score
```

######

This version uses a Boolean flag to control the loop. This time we are using a flag called `erroneous` which is `True` at the start of the loop, and is set to `False` if the input is acceptable. You must choose the clearest implementation that solves the problem.

### Programming validation with a while loop

### Worked Example 1: Validating menu choices

In this example, which is part of a shopping program, we'll use `while` to check the choice entered is in the range 1 to 4

```python
def getChoice():
  print("Menu:")
  print("1. Add item")
  print("2. List items")
  print("3. Delete items")
  print("4. Go back")
  choice = input("Enter a choice between 1 and 4: ")
  while choice < 1 or choice > 4:
      print("Invalid input. Please try again.")
      choice = input("Enter a choice between 1 and 4: ")
  return choice
```

### Worked Example 2: Validating an input day of the week

Imagine we are writing an appointment booking system. We need a function to input a day of the week. In this function, an input string can be one of the days of the working week in short form ("Mon", "Tue" up to "Fri"). This time we will use a Boolean flag `looping` to control the loop. Once `looping` gets set to `False`, the loop will terminate on next iteration.

> Note that this example uses a **list** to store valid day names. We will look at lists in Unit 6.

```python
def getDay():
    looping = True
    while looping:  
        day = input("What day of the week are you available? ")
        if day in ["Mon", "Tue", "Wed", "Thu", "Fri"]:
            looping = False
        else:
            print("Invalid input, try again!)
    return day
```

### Summing numbers with a while loop

Imagine you want to add up a stream of values: test scores, daily temperatures, the prices of goods in your shopping basket. We can write a `while` loop to repeatedly input these values and sum them. The code that performs this activity is so common it has a name: 

### The accumulator pattern

Let's write a function that uses a `while` loop to repeatedly input integers that represent test scores, sum these scores and return the total. We will terminate the loop when the user enters `-1` (otherwise it will run forever!)  

```python
def sumScores():
    total = 0
    score = 0
    while score != -1:
        score = int(input("Enter a score: "))
        if score != -1:
            total = total + score
    return total
```

You can see that if the user enters the special value `-1` the loop ends next time it iterates. If the input value is not `-1` it is treated as a data value, and added to the total. 

> *Fun fact:* You don't need to know this for GCSE, but a special value that terminates a loop in this way is called a **sentinel** value, and this type of loop is often called a **listener** loop because it listens for the sentinel value!

## Boolean flag-controlled version

We can make the above code more readable if we rewrite (sometimes called *refactor*) our code to use a Boolean flag to control iteration:

```python
def sumScores():
    total = 0
    finished = False
    while not finished:
        score = int(input("Enter a score: "))
        if score == -1:
            finished = True
        else:
            total = total + score
    return total
```
