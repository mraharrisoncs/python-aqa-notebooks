### Using compound expressions

So far, we have used simple Boolean expressions in our while loops: `len(password) <-10` and `passwordRepeat != password`. But we can use a compound Boolean expression, one that combines multiple conditions using logical operators such as `and`, `or`, and `not`. This can make our loops more powerful and flexible.

## Compound expressions

Let's revisit the `getPassword` program and enhance it to ensure that the password not only meets a minimum length requirement but also **starts with an uppercase letter** and **contains a hash symbol,** **`#`.**

Here is the enhanced version of the function:

```python
def getPassword():
    # input password repeatedly until it meets all criteria
    password = ""
    while len(password) < 10 or not password[0].isupper() or '#' not in password:
        password = input("Type new password, min length 10," \
                          + " first uppercase, include #: ")
      
    # input password again until it matches the first password
    passwordRepeat = ""
    while passwordRepeat != password:
        passwordRepeat = input("Type your new password again: ")
    
    # return the new password
    return password
```

#### Explanation

Let's look at that first compound condition in detail. It contains three separate conditions:

* `len(password) < 10` - we met this earlier, this will be `True` if the password is too short.

* `not password[0].isupper()` - this grabs the first character of the password with `password[0]` then uses the `isupper()` string method, which returns `True` if this is uppercase. Finally we negate this result using `not`, so this condition will be `True` if the first character is **not** an uppercase letter - another indicator that our password is invalid.

* `'#' not in password` - this uses the string comparison operator `not in` (see Unit 3) which returns `True` when a substring is **not in** a larger string, in this case when there is no `#` symbol in our password - the final indicator for an invalid password.

These three conditions are connected with Boolean `or` operators. That's because if any one of them is `True`, our password is too weak and must be rejected, and the `while` loop will continue running. Only when all three conditions are `False` will our loop terminate.

> ## Negative thinking
>
> While loops are tricky to get your head around sometimes, just remember they keep looping **while** a condition is **True**. Sometimes it's more intuitive to state the  **terminating** condition, i.e. what needs to happen for the loop to **end**. If this is the case, don't forget to **negate** that condition before using it on your **while** statement!

````python
# password2.py
def getPassword():
    # input password repeatedly until it meets all criteria
    password = ""
    while len(password) < 10 or not password[0].isupper() or '#' not in password:
        password = input("Type new password, min length 10," \
                          + " first uppercase, include #: ")
    
    # input password again until it matches the first password
    passwordRepeat = ""
    while passwordRepeat != password:
        passwordRepeat = input("Type your new password again: ")
    
    # return the new password
    return password

# MAIN PROGRAM
print("Let's set your password!")
password = getPassword()
print(f"your new password is {password}")
````

### More while loops with compound expressions

Let's now write a                ` checkPassword` program which inputs a password attempt, and checks against the password variable, returning `True` if correct within three attempts, before returning `False`. For simplicity we have set the password to "Computer#Science", but in reality we would retrieve a password from an encrypted database!

```python
def checkPassword():
    password = "Computer#Science"
    attempts = 3
    passwordAttempt = ""

    while passwordAttempt != password and attempts > 0:
        passwordAttempt = input(f"Type your password, {attempts} attempts remaining: ")
        attempts -= 1
 
    if passwordAttempt == password:
        return True
    else:
        return False
```

We are using an `if` statement after the while loop ends, to check the last attempt, and decide whether to return `True` or `False`, because we cannot be sure if the loop has terminated because the last password attempt was correct, or the attempts ran out!

## Returning a boolean directly

There's actually no need to use an `if` here, we can just return the result of the comparison like this:

```python
return passwordAttempt == password:
```

This is because the expression `password_attempt == password` evaluates to `True` if the last password attempt was correct, and `False` if not, so we can just return that value directly.

## Final version of check\_password()


### Using Boolean flags to control iteration

Remember our password strength checker? It ended up with a quite messy condition that looked like this:

```python
while len(password) < 10 or not password[0].isupper() or '#' not in password:
```

When your while loop condition starts to grow like this, it might be time to replace it with a Boolean **flag** which controls the iteration. Then you can test the conditions and alter the flag within the loop body, like this *(I've trimmed the code by taking out the check for repeat password)* :

```python
def getPassword():
    # input password repeatedly until it meets all criteria
    valid = False
    while not valid:
        password = input("Type new password, min length 10," \
                              + " first upper, include #: ")
        valid = True # assume input is valid, and then test assumptions
        if len(password) < 10 or not password[0].isupper() or '#' not in password:
            valid = False # password is weak so loop should continue
    return password
```

We are now controlling the loop with a Boolean flag called `valid`. After inputting the new password, this flag is set to `True`, we check for password strength. If the checks fail, the `valid` flag is changed back to `False`.  Note that if the `valid` flag remains `True` the while loop will terminate, because its condition `not valid` will be `False`. 

> ### Python reads like English!
>
> Tip: If all the `True` and `False`s are confusing you, just remember, Python was designed to be very readable. Each line of code says what it means. So `while not valid` means "keep looping while my password is not valid"! Choose an identifier for your Boolean flag variable that makes sense, and your code will be easy to follow.

## Adding complexity

Now we have a Boolean flag controlling our while loop, it's easy to add complexity to our loop control decisions. We can just write more lines of code that alter the `valid` variable as we need them. Let's say we now want to check that the password is not in a **blocklist** of banned passwords. Let's add that code now:

```python
def getPassword():
    blocklist = ["ABCDEFGHIJ#", "QWERTYUIOP#", "A123456789#"]
    # input password repeatedly until it meets all criteria
    valid = False
    while not valid:
        password = input("Type new password, min length 10," \
                          + " first upper, include #: ")
        valid = True # assume input is valid, and then test assumptions
        if len(password) < 10 or not password[0].isupper() or '#' not in password:
            valid = False # password is weak so loop should continue
        if password in blocklist:
            valid = False # password is in block list so loop should continue
    return password
```

We just add the new check for 'presence in the blocklist' as a separate `if` statement. It sets the `valid` flag back to `False` if the password is found in the blocklist. Because we are using a single Boolean variable to control the loop., we could keep adding new password strength checks without the while loop condition becoming long and complicated!

> Tip: there are many ways to solve this problem, the code above is just one way! Find the most elegant way of solving the problem: concise but readable, and easy to debug.

````python
# get_password_flag.py
def getPassword():
    blocklist = ["ABCDEFGHIJ#", "QWERTYUIOP#", "A123456789#"]
    # input password repeatedly until it meets all criteria
    valid = False
    while not valid:
        password = input("Type new password, min length 10," \
                          + " first upper, include #: ")
        valid = True # assume input is valid, and then test assumptions
        if len(password) < 10 or not password[0].isupper() or '#' not in password:
            valid = False # password is weak so loop should continue
        if password in blocklist:
            valid = False # password is in block list so loop should continue

    # return the new password
    return password

# MAIN PROGRAM
print("Let's set your password!")
password = getPassword()
print("your new password is", password)
````
