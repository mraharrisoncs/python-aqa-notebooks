### Pseudocode WHILE ENDWHILE and REPEAT UNTIL

## Pre-conditioned loops in pseudocode 

As we have seen previously, AQA exam papers and reference material use both pseudocode and Python. 

When you see pseudocode `while` loops in AQA exam papers we will use keywords `WHILE` and `ENDWHILE `like this:

```python
WHILE BoolExp
  # statements here
ENDWHILE
```

for example, our old password length validation loop looks like this:

```python
password ← ""
WHILE LEN(password) < 10
    password ← USERINPUT
ENDWHILE
```

A `while` loop always tests the condition **at the start** of each iteration. So in the above example, the password length is checked on entry to the loop, and again on every iteration. 

This is called a **pre-conditioned** loop. Some programming languages - but not Python - allow you to check the loop condition **at the end** of the loop. This is called a **post-conditioned** loop. 

## Post-conditioned loops - REPEAT UNTIL

These loops are often better suited to a problem than pre-conditioned loops. AQA pseudocode keywords `REPEAT` and `UNTIL` can be used to create a post-conditioned loop. Here is the password checker loop using these keywords:

```python
REPEAT
    password ← USERINPUT
UNTIL LEN(password) >= 10
```

This code is simpler and more elegant than the original WHILE version: we don't need to initialise the `password` variable before we enter the loop, and the logic feels more natural: "repeatedly input a password until it's length is at least 10".

We can't do this in Python, but here is a sample of **Lua** - a beginner-friendly language used in game development such as in Defold and Roblox - that implements this post-conditioned loop:

```
# Lua code 
repeat
    password = io.read()
until string.len(password) >= 10
```

## "Faking" post-conditioned loops in Python

We don't have native `REPEAT-UNTIL` in Python but we can "fake it" with `while`. Here is our password checker loop using only while:

```python
password = ""
while len(password) < 10:
    password = input("Enter a password: ")

```

Here is another example of a `REPEAT-UNTIL` in pseudocode, this program repeatedly sums coin values until a price is reached:

```python
price = 150
total = 0
REPEAT
    coin ← USERINPUT
    total ← total + coin
UNTIL total >= price
```

Here is how we can implement that in Python without `REPEAT-UNTIL`, by using a Boolean variable to control the loop, and changing the `UNTIL` to an `if` statement:

```python
# Python 'fake' post-conditioned loop
price = 150
total = 0
looping = True
while looping:
    coin = int(input())
    total = total + coin
    if total >= price:
        looping = False 
# rest of program
```

### 🍵 Example 1: Vending Machine Coin Counter

Here's a simple function that uses the while loop we saw earlier to input coin values until a total is reached or exceeded. The function returns the change due.

```python
def inputCoins(price):
    total = 0
    looping = True
    while looping:
        print("credit so far =", total)
        coin = int(input("Enter a coin: "))
        total = total + coin
        if total >= price:
            looping = False 
    return total - price

# main program
change = inputCoins(450)   # change 450 to test with other prices
print("Your change is", change)
 
```

#### Explanation:

* The function `input_coins()` takes one parameter `price` which will be the price of an item

* We initialise the `total` to zero, and the loop control Boolean flag `looping` to `True`

* We now enter the loop, and inside the loop we output the running total, and input a new coin. If the total reaches the price, or goes over it, the loop control variable is set to False and the loop ends.

* We return the difference between the total and the price - this is the change due

* The main program calls the function with the price **450** (£4.50), and saves the return value as `change` and then outputs this.

***

### 🌡️ Example 2: Temperature analysis  

Here's a function that inputs temperatures repeatedly, until the user types "99". It then returns the average temperature.

```python
def analyseTemps():
    total = 0
    count = 0
    finished = False
    while not finished:
        temp = int(input("Enter a temperature in C (99 to end): "))
            while temp < -40 or temp > 99:
                temp = int(input("Invalid, retry: "))
        if temp == 99:
            finished = True
        else:
            total = total + temp
            count = count + 1
    return total / count

# main program
print("Average temperature was:", analyseTemps())
```

#### Explanation:

* The function `analyse_temps()` initializes two variables, `total` and `count`, to zero and a Boolean flag `finished` to `False`, and then enters a `while` loop that continues until `finished` is `True`.

* Inside the loop, it prompts the user to input a temperature in Celsius, with "99" as the 'sentinel' value to end the input.

* If the input temperature is outside the valid range (-40 to 99), it prompts the user to retry until a valid temperature is entered.

* If the input temperature is "99", the loop ends by setting `finished` to `True`.

* Otherwise, it adds the temperature to `total` and increments `count` by 1.

* The function calculates the average temperature using `total/count` and returns this.

* The main program calls the function and outputs the result.

````python
# word_game_solution.py
def wordScore():
    # initialise total score 
    totalScore = 0
    # initialise Boolean flag (if using)
    done = False
    while not done:
        # input letter
        letter = input("Enter a letter: ")
        # check for sentinel
        if letter == ".":
            done = True
        # calculate letter score
        else:
            if letter in "aeiou":
                score = 1
            elif letter in "qxz":
                score = 3
            else:
                score = 2
        # add to total score
            totalScore += score
    # return total score
    return totalScore

# main program
print("Enter a word, letter by letter,")
print("full stop to finish.")
print(f"Your word scored: {wordScore()}")
````

````python
# word_game_solution_structured.py
# New version with separate functions
# for input & score calculation and main loop

def getLetter():
    # input letter and score it
    letter = input("Enter a letter: ")
    if letter == ".":
        return -1 # sentinel value
    elif letter in "aeiou":
        return 1
    elif letter in "qxz":
        return 3
    else:
        return 2
    
def wordScore():
    # calculate word score
    totalScore = 0
    done = False
    while not done:
        score = getLetter()
        if score == -1: # input was "."
            done = True
        else:
            totalScore += score
    return totalScore

# main program
print("Enter a word, letter by letter,")
print("full stop to finish.")
print(f"Your word scored: {wordScore()}")
````

### Infinite loops and beyond!

A big issue with condition-controlled iteration is the risk of an **infinite loop**, a loop that never ends!

### Accidental infinite loops

Remember our countdown program? This slightly altered version - which should count down from 10 in steps of two - contains a bug, can you spot it? Try running it!

Hint: you might need to hit the stop button - the red square that looks like this 🟥 at the top right of the screen! 

```python
def countdown():
  # count down from 10 in twos
  counter = 10
  while counter != 1:
    print(counter, "!")
    counter -= 2
  print("lift-off !")

# MAIN PROGRAM
countdown()
```

`while counter != 1` says "keep looping while the counter is **not equal to 1**", or in other words, "stop looping when the counter reaches 1".

However, `counter` begins at 10 and counts down in twos, so it will go `8, 4, 2, 0, -2...` ! It will never equal 1.

We have coded an **infinite loop**. Can you work out how to fix it?  

### While True

Sometimes the simplest way to code a while loop is to make it loop "forever", and **break out** of the loop when something happens. 

We can code any **Boolean expression** that evaluates to `True` on our while loop condition, to make it loop forever. Here are some examples:

```python
while counter == counter:
while 1 == 1:
while 1 < 2:
while "x" = "x"
```

All of these expressions evaluate to `True` at all times, and the while loop will run forever. But there is a much simpler expression that is always `True` - the keyword `True` itself!

```python
while True:
  # this code block will run forever
```

This is how we deliberately code an 'infinite' loop: with `while True`. But now we must find another way of terminating the loop, otherwise it will loop forever!

***

### Beyond the spec - breaking out!

In AQA GCSE exam questions, your **indefinite loops** should use a terminating condition - you will never need to use `while True`. However if you find yourself coding it in Python, you'll need to know how to break out...

### Returning out of a loop

There are two ways we can break out of a while loop, and we must use one of them inside a `while True` loop, or it will never end! First we can **return** out of the function that the while loop is part of. This works fine, if the function does not need to do anything else.

 

Here is a validation example from Lesson 3, showing this technique.

```python
def getScore():
  score = int(input("Enter a score between 1 and 100: "))
  while True:
      if score < 1 or score > 100:
          print("Invalid input. Please try again.")
          score = int(input("Enter a score between 1 and 100: "))
      else:
          return score # return out of the loop 
```

### The **break** keyword

Or we could use the Python keyword `break` which says "exit the current loop". 

```python
def getScore():
  score = int(input("Enter a score between 1 and 100: "))
  while True:
      if score < 1 or score > 100:
          print("Invalid input. Please try again.")
          score = int(input("Enter a score between 1 and 100: "))
      else:
          break # break out of the loop
  return score 
```

> These techniques are beyond the AQA specification, so will not be expected in exams, but they are valid if used.

````python
# q1_solution.py
'''
Print a countdown from the given number down to 1,
then print 'Lift off!'.
'''
def countdown(start):
    while start > 0:
        print(start)
        start -= 1
    print("Lift off!")

# main program
countdown(5)
countdown(10)
````

````python
# q2_solution.py
'''
Repeatedly ask the user for their age until they enter
a valid value between 0 and 120. Then print 'Age accepted'.
'''
def getValidAge():
    age = int(input("Enter your age: "))
    while age < 0 or age > 120:
        age = int(input("Invalid age. Enter your age: "))
    print("Age accepted")

# main program
getValidAge()
````

````python
# q3_solution.py
'''
Write a function 'getPrice' that returns the price of an item:
"cola" → 2.00, "crisps" → 1.20, "water" → 1.00
Then write the 'runMachine' function that repeatedly
inputs items and calls 'getPrice' to get the price of the item
until the user types 'quit', and then outputs the total cost.
'''
def getPrice(item):
    if item == "cola":
        return 2.00
    elif item == "crisps":
        return 1.20
    elif item == "water":
        return 1.00
    else:
        return 0

def runMachine():
    total = 0
    item = input("Enter an item (or 'quit' to finish): ")

    while item != "quit":
        total += getPrice(item)
        item = input("Enter an item (or 'quit' to finish): ")

    print("Total: £", total)

# main program
runMachine()
````
