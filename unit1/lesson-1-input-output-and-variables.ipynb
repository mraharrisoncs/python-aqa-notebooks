### Variables in Python

Let's take a look at **variables** in Python, and how they differ from **literals**.

### Literals

So far we have just printed **literals**. As their name suggests, when you print **literals**, Python just outputs **literally** what you give it.

Anything inside a pair of quotes, such as `"Hello World!"` is a **string literal**, and a number like `5` or `3.14` is a **numeric literal**.

```python
>>> print("Hello World!")
Hello World!

>>> print("pi approximates to", 3.14)
pi approximates to 3.14
```

This is not very useful, as we're only displaying things typed in by the programmer. We really need a way of storing things and using them later. That's where **variables** come in.

### Variables are names for things

In programming, a **variable** is a **named storage location in memory** that holds a value. Think of it as a container where you can store data that you want to use later in your program. Variables allow you to label and manipulate data easily.

### Assigning Variables

To create a variable in Python, you use the **assignment** operator **`=`** like this:

```python
variableName = value
```

* `variableName` is the name you give to the variable, also called its **identifier**.

* `value` is the data you want to store in the variable.

For example, let's create a variable named `age` and assign it the value `16`:

```python
age = 16
```

Here, `age` is the variable name, and `16` is the value stored in it.

### Printing Variables

You can use the `print()` function to display the value of a variable. Let's print the `age` variable:

```python
>>> age = 16
>>> print(age)
16
```

### Combining Variables and Literals

You can also print variables alongside other text (string literals) by including them in the `print()` function. For example:

```python
>>> age = 16
>>> name = "Alice"
>>> print("Hello, my name is", name, "and I am", age, "years old.")
Hello, my name is Alice and I am 16 years old
```

This will output: `Hello, my name is Alice and I am 16 years old.`

By using variables, you can easily change the data in your program without having to rewrite the entire code.

### Input using input()

Let's now gather input from users in Python using the `input()` function. Input is really useful, as it allows your program to **respond to user actions**.

### Understanding input()

The `input()` function in Python will pause and wait for the user to type something and press Enter. If we **assign** this input data to a **variable** we can use it later in our program, like this:

```python
print("Enter your name:")
name = input()
print("Hello", name, "!")
```

In this example:

* `print("Enter your name:")` displays a message prompting the user to enter their name.

* `name = input()` waits for the user to type their name and press Enter. This name is stored in the variable `name`.

* Finally, `print("Hello, ",  name, "!")` outputs a greeting using the name provided by the user.

### Casting functions

Sometimes, you need to convert data from one type to another. For example, `input()` always returns a **string** so we need to cast this with `int()` or `float()` if we want to save it as a number. Python provides **casting functions** to do this.

### Casting functions

**`int()`**: Converts to an integer🔢:

```python
age = int("25")  # Converts string to integer
year = int(input("What year were you born? ")) # string to int
```

**`float()`**: Converts to a float **🧮**:

```python
temperature = float("36.6")  # Converts string to float
weight = float(input("Enter weight in kg: "))  # string to float
```

**`str()`**: Converts to a string📜:

```python
ageStr = str(25)  # Converts integer to string
print("Your weight is: " + str(weight)) # float to string
```

### About data types

When we stored the user's name earlier, we used a variable called **name**. Because the name contains a string of characters, we call this a **string** variable. Their age was a whole number, and this is called an **integer.** Together, these different kinds of data are called **data types.** 

In Python, **data types** are essential for defining the kind of data a variable can hold. There are four main data types in Python, known as the **primitive** data types. These are:

#### 🔢Integer 🧮Float 📜String ✅/❌Boolean

Let's explore these data types in more detail.

### 🔢 Integer

An integer is a **whole number** without a decimal point. It can be positive, negative, or zero. Integers are used in situations where fractional values are not needed, such as counting items or indexing:

```python
age = 25
numberOfStudents = 30
```

### **🧮** Float

A float, or real number, is a number that **includes a decimal point**. Floats are used when precision is required, such as in scientific calculations or when dealing with currency, weights and measures:

```python
temperature = 36.6
price = 19.99
```

### 📜 String

A string is a **sequence of characters**. Strings store text such as names, addresses, post codes, anything that might contain letters, numbers and symbols.

```python
name = "Alice"
greeting = "Hello World!"
postcode = "DE6 2PZ"
```

### ✅/❌Boolean

A Boolean represents **one of two values:** **`True`** **or** **`False`**. Booleans are used in conditional statements and logical operations:

```python
isStudent = True
hasPassed = False
```

### Worked Example 1: BMI calculator

Your BMI (body mass index) is calculated from your weight in kilograms divided by your height in metres squared. Here is the code, also in **bmi.py** attached.
