### What is a subroutine?

A **subroutine** is a **named  block of code** that may be executed (called) by simply writing its name in a program statement. We sometimes call it an 'out-of-line' block of code, meaning it is not 'in-line' with the rest of the program.

Here is the "Hello World!" code rewritten as a subroutine.

```python
def helloWorld():
    print("Hello World!")
```

This gives the `print("Hello World!")` statement the  **identifier** of `helloWorld`. Any time we want this subroutine to run and output "Hello World!" we just **call** it by using its identifier, like this:

```python
helloWorld()
```

Notice that we write the identifier of the subroutine followed immediately by a pair of brackets.  The pair of brackets, `()`, tells Python that this identifier is the name of a **subroutine**, not a variable. 

When Python sees `helloWorld()` it will run the code defined earlier by the `def helloWorld()` instruction.

### Indentation, colons and code blocks!

Remember our `helloWorld()` subroutine? The **subroutine definition** ended in a **colon** **`:`** and the line of code following this colon was **indented** - this means we typed four spaces in front of the keyword `print`. 

The colon says to Python "here follows a **code block**" and the **indentation** tells Python which lines belong to the **code block**. 

```python
def helloWorld():
    print("Hello World!")

# main program
helloWorld()
```

 

Once the indentation **stops**, Python knows that's the **end of the code block** that followed the colon, and therefore the end of the subroutine definition. 

## Indented code blocks

A code block have lots of lines of code, but they all have to be **indented** by the same number of spaces. 

> **Tip:** we usually indent by four spaces, but sometimes you will see two spaces used. As long as you're consistent, and indented statements all line up vertically, the number of spaces doesn't matter.

Here is another example, with a **code block** of three lines of code inside the subroutine:

```python
def helloEveryone():
    print("Hello Earth!")
    print("Hello Venus!")
    print("Hello Mars!")
helloEveryone()
```

This subroutine contains a code block of three `print()` statements. Because they are all **indented** Python knows they are a single code block, and therefore belong to the subroutine. 

> Note: I've deleted the whitespace line, and the comment line, to show that these do NOT matter, the subroutine definition ends **as soon as Python sees a line without indentation.**

### Why subroutines?

Subroutines help us organize programs into chunks that match how we think about the solution to the problem.

## Decomposing a problem

 Programming is problem-solving, and the first step is **decomposition.**

If we **decompose** a problem into sub-problems, we can then write a **subroutine** to solve each sub-problem. These subroutines come together to make a larger **program** which solves the big problem.

Let's revisit the "chatbot" program:

```python
name = input("What is your name? ")
print("Hello, ", name, "!")
subject = input("What is your favourite subject")

```

To demonstrate subroutines we will break this program into chunks, one to ask the users name and say hello, and another to ask their favourite subject and respond to that.

### Python `def` keyword

We define a subroutine in Python with the **def** keyword, followed by a name for the subroutine. Once we have defined these subroutines we need to ***call*** them in the main program like this:

```python
def sayHello():
  name = input("What is your name? ")
  print("Hello, ", name, "!")

def favSubject():
  subject = input("What is your favourite subject")
  print(subject, "is a great subject!")

# main program
sayHello()
favSubject()
```

### Explanation

Remember that `def` creates and stores a subroutine for use **later**, it does **not** execute straight away. The indented code block below the `def` statement is the subroutine **body** which executes when the subroutine is called. As soon as we stop indenting,  the definition is complete. 

The subroutine body - the code block under the `def` statement - will run only when it is called by name. For example, `say_hello()` will run the subroutine defined previously with `def say_hello():`

## Another example

If we want to draw a square with asterisks, we could do this:

```python
print("******")
print("*     ")
print("*     ")
print("*     ")
print("*     ")
print("******")
```

But if we use the **computational thinking** skill of **pattern matching** to notice that the first and last lines are identical, and also that the middle four lines are identical, we could instead write two subroutines, and call them repeatedly, like this:

```python
def horizontal():
  print("******")

def vertical():
  print("*    *")

# main program
horizontal()
vertical()
vertical()
vertical()
vertical()
horizontal()
```

### Explanation

The first subroutine `horizontal()` defines how to draw a horizontal line of asterisks. The second subroutine `vertical()` defines how to draw one pair of asterisks to represent one section of the vertical lines. In the main program we call these two subroutines repeatedly to draw the square.

## Link to Abstraction

When we hide the details of how something works, this is **abstraction**. Subroutines allow us to squirrel away code behind a meaningful identifier, so we don't have to think about it any more! 

In this program the details of how to print the horizontal and vertical edges have been **abstracted** away from the main program into the subroutines. 

This might not save any lines of code in this example but:

* it makes the **main program** very simple to write and easy to understand - we have **abstracted away** the details into the subroutines.

* it allows us to change the shape of the figure easily by editing the subroutines, for example we can make the shape wider by just adding asterisks or spaces to a single print statement, while the main program stays the same.

* in a later unit we will see how to repeat lines of code, so we don't need to code `vertical()` multiple times in our program, saving lots of coding.

### Procedures and functions

To recap, a **subroutine**, also called a **subprogram** is any block of code with a name or **identifier** that performs a single process within a larger program.

There are two types of subroutines: **procedures** and **functions**. A **function** **returns** a value, while a **procedure** does not, but otherwise they are the same.

## Procedures

The subroutines in our previous example were actually **procedures** because they did not return a value:

```python
def sayHello():
  name = input("What is your name? ")
  print("Hello", name, "!")

def favSubject():
  subject = input("What is your favourite subject")
  print(subject, "is a great subject!")

# main program
sayHello()
favSubject()
```

You can try out this code in the attached **proc.py**

## Functions

A function is a subroutine that returns a value. This means that as well as performing a small process, the function passes  a value back to the calling instruction, like this:

```python
def getName():
  ans = input("What is your name? ")
  return ans

# main program
name = getName()
print("Hello", name, "!")
```

### Explanation

* We define, `get_name()` as a subroutine just like before, but inside the **body** of the function (the indented block), we see a `return` statement, meaning this is a **function**.

* In the main program, `get_name()`  **calls** the function. The function then executes

* Inside the function, we input the user's name and assign it to a variable called `ans`.

* The function passes this value back through the `return ans` statement. 

* In the main program, `get_name()` becomes this value, which is then stored in the variable `name`.

* The `print` statement prints the greeting, using the `name` variable again.

You can try out this code in the attached **func.py**.

## Everything is a function in Python!

To simplify conversations about subroutines, procedures and functions, and because of the way Python 3 was designed, Python programmers don't make the distinction between procedures and functions at all! 

Every subroutine is called a **function** in Python, whether it returns a value or not. From now on, whenever we create a subroutine in Python with the `def` keyword, we will call it a **function**, as all Python programmers do.

Be careful, though, exam questions will need you to know the difference!

## Generalisation using parameters

We could make a general purpose function called `greet()` that outputs any greeting **passed** to it as a **parameter**. Like this:

```python

# main program

greet("Witaj świecie!")  # "Witaj świecie!" is passed into the parameter
```

### Structured programming

**Structured programming** is a method of designing and writing programs that are easy to understand, debug, and maintain, by using a **modular** approach. This means breaking down a program into **modules** - smaller, manageable parts - and coding these separately. In Python we use **functions** to help us write modular code.

### Why Use structured programming?

1. **Improved Readability**: By dividing a program into functions, each part is easier to understand, making the overall program easier to read.

2. **Easier Debugging**: Smaller sections of code are easier to test and debug. Testing can focus on a specific subroutine rather than the entire program.

3. **Reusability**: Subroutines can be reused in different parts of the program or even in other programs, saving time and effort.

4. **Simplified Maintenance**: We can change individual subroutines without affecting the rest of the program, making maintenance simpler.

## Designing a structured program

Imagine we want a program to help a mathematician working on circles. Given the radius of a circle, the mathematician needs to calculate its diameter, circumference and area. We would **decompose** this problem into sub-problems, one problem for each of the outputs needed (diameter, circumference and area).

When we are **decomposing** a problem in this way, we sometimes plan the program using a diagram like this, called a **top-down design**. The whole program is at the top, and underneath are the sub problems we have identified:

### Example in Python

Now we can code this in Python, by writing a function for each sub-problem, and then a main program that inputs a radius and calls the functions:

```python
PI = 3.1415

# function to calculate the diameter of a circle
def diameter(radius):
    return 2 * radius

# function to calculate the circumference of a circle
def circumference(radius):
    return 2 * PI * radius

# function to calculate the area of a circle
def area(radius):
    return PI * radius ** 2


# main program
radius = float(input("Enter the radius: "))
print("Diameter: ", diameter(radius))
print("Circumference: ", circumference(radius))
print("Area: ", area(radius))
```

### Explanation

* We define a variable `PI` to be `3.1415`, an approximation of pi. 

  > We use capitals because we are never going to change this variable. `PI` is acting as a **constant** which we will discuss on the next page.

* We define a function `area` that takes the `radius` of a circle as a parameter and calculates and returns the area.

* We define a function `circumference` to take  `radius` and calculates and return the circumference.

*   Then the main program prompts the user to input a radius, then calls these functions to compute and print the area and circumference. 

Now answer these questions to check your understanding.

````python
# geometry.py
PI = 3.1415

# function to calculate the diameter of a circle
def diameter(radius):
    return 2 * radius

# function to calculate the circumference of a circle
def circumference(radius):
    return 2 * PI * radius

# function to calculate the area of a circle
def area(radius):
    return PI * radius ** 2


# main program
radius = float(input("Enter the radius: "))
print("Diameter: ", diameter(radius))
print("Circumference: ", circumference(radius))
print("Area: ", area(radius))
````

### Writing a quiz game

A maths teacher wants a game to quiz students on their times tables. The game should display a question in the form of a multiplication problem, such as `4 * 8` or `9 * 3`, using random integers between 1 and 10 inclusive. 

The user inputs their answer and the program checks the answer and adds 1 to their score if correct. The quiz should ask three questions and then give a score out of three and stop. 

We can decompose this problem into these sub-problems:

* generate a question in the form `a * b` where a and b are random integers between 1 and 10, and display it

* ask for the user's answer

* check the answer and add 1 to score if correct

* repeat 2 more times

* output the final score.

>  

You can start by writing the **main program** section like this, just by giving reasonable **identifiers** (function names) to the list of sub-problems we wrote above:

Note how we have not yet defined **how** to do any of these processes. But we have written the outline structure of our program. We have a high-level **abstraction** of the program required.
