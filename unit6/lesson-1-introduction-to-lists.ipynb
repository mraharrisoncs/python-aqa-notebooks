### Why use lists?

### Why use lists?

Remember our temperature program? What if - instead of just calculating a total or average and discarding our input data - we wanted to store all the input data for future use? We *could* store each day's temperature in a separate variable like this:

```python
temp1 = int(input("Enter temperature 1: "))
temp2 = int(input("Enter temperature 2: "))
temp3 = int(input("Enter temperature 3: "))
temp4 = int(input("Enter temperature 4: "))
''' continued
```

But this would need one line of code and **one integer variable for each value**, and if we wanted to store the temperature *every hour for several years*, that's a lot of code and a lot of variables! 

Also, working with this data would need yet more clunky code, for example to sum the first seven values:

```python
>>> total = temp1 + temp2 + temp3 + temp4 + temp5 + temp6 + temp7
```

## Creating and indexing a list

Instead we can create a **list** called `temps` and store **all** our temperatures under this **identifier.** This is how we do that:

```python
>>> temps = [24, 26, 31, 26, 21, 18, 22]
```

We can now access any **item** in the list as long as we know its **position** in the list, which we call the **index** of the item. But remember from string indexing, Python always counts from **zero**. For example:

```python
>>> temps = [24, 26, 31, 26, 21, 18, 22]
>>> temps[0]
24
>>> temps[5]
18
```

Now we can use a really cool function called `sum()` to add up all the values!

```python
>>> temps = [24, 26, 31, 26, 21, 18, 22]
>>> sum(temps)
168
```

And we can use `len()` - just like we did in string handling - to find the length of the list.

```python
>>> temps = [24, 26, 31, 26, 21, 18, 22]
>>> len(temps)
7
```

So now we can find the **average** very easily!

```python
>>> temps = [24, 26, 31, 26, 21, 18, 22]
>>> sum(temps)/len(temps)
24.0
```

## About lists

We chose the identifier `temps` because the list stores temperatures, making `temps` a **meaningful** name for the list. Your list identifier should say what is in the list. It's also a good idea to make a list identifier **plural** (e.g. `temps`, `scores`, `days_of_the_week`) to make clear it contains multiple items.

The list is an example of a **data structure** - a way of storing and organising a collection of data under a single identifier, so it can be quickly and easily processed in many ways.

### List slicing

You've seen how to access individual elements in a list using their index positions. What if we want a sequence of more than one element, e.g. the first three or the last two?

## List slicing

The basic Python syntax for **list slicing**, is this:

```python
identifier[start : end]
```

where **identifier** is our variable name, **start** is the starting position of the sequence, and **end** is the first element we **don't want!** 

> *Tip:* The syntax we use for list slicing syntax is identical to **string slicing** syntax. All the tricks we learned in **string slicing** can be used with lists too!

Just like in string slicing, we must make the second value **one more** than the last element we want to extract.

### Structure of a list

Here is that **fruits** list again, showing the index of each element:

```
fruits
+--------+--------+--------+--------+--------+--------+
| index  |   0    |   1    |   2    |   3    |   4    |
+--------+--------+--------+--------+--------+--------+
| value  | apple  | orange |  kiwi  | banana |  pear  |  
+--------+--------+--------+--------+--------+--------+
```

### Indexing from the start

We've seen already we can access individual elements by their index position, counting from zero:

```python
>>> fruits = ["apple", "orange", "kiwi", "banana", "pear"]
>>> fruits[2]
'kiwi'
```

### Indexing from the end

We can also count backwards **from the end of the list** using negative numbers. Index `-1` indicates the last element, `-2` the second last, and so on:

```python
>>> fruits = ["apple", "orange", "kiwi", "banana", "pear"]
>>> fruits[-1]
'pear'

>>> fruits[-2]
'banana'

```

Here we can see the negative indexes, underneath the usual positive indexes:

```
fruits
+----------+--------+--------+--------+--------+--------+
| index    |   0    |   1    |   2    |   3    |   4    |
+----------+--------+--------+--------+--------+--------+
| negative |   -5   |   -4   |   -3   |   -2   |   -1   |
+----------+--------+--------+--------+--------+--------+
| value    | apple  | orange |  kiwi  | banana |  pear  |  
+----------+--------+--------+--------+--------+--------+
```

### List slicing

We can also give both **start** and **end** values, to retrieve a **list slice**. This behaves in the same way as **string slicing** (see Unit 3), we give the **start** index, and the index of the first element we **do not want**, as the **end** index. Here is how that works:

```python
>>> fruits = ["apple", "orange", "kiwi", "banana", "pear"]
>>> fruits[0:2]
['apple', 'orange']

>>> fruits[2:5]
['kiwi', 'banana', 'pear']

```

If we leave either **start** or **end** value blank, Python will assume we mean the start or end of the list:

```python
>>> fruits = ["apple", "orange", "kiwi", "banana", "pear"]
>>> fruits[:2]
['apple', 'orange']

>>> fruits[3:]
['banana', 'pear']

```

We can use negative numbers in our slice too, here's how to chop off the final element:

```python
>>> fruits[:-1]
['apple', 'orange', 'kiwi', 'banana']
```

### Append, change and remove list items

### Append to a list

Lists can grow and shrink: they are not fixed in size. We can add items to the end of the list using the `append` method. Here is an example:

```python
>>> fruits = ["apple", "orange", "kiwi", "banana", "pear"]
>>> fruits.append("plum")
>>> fruits
['apple', 'orange', 'kiwi', 'banana', 'pear', 'plum']
```

### Change list element

We can change or **overwrite** a list element by using its index, like this:

```python
>>> fruits[1] = "peach"
>>> fruits
['apple', 'peach', 'kiwi', 'banana', 'pear', 'plum']
```

You can see that element with index 1 has changed from "orange" to "peach".

### Remove list element

The `remove` method can find and remove a list item by name:

```python
>>> fruits = ['apple', 'peach', 'kiwi', 'banana', 'pear', 'plum']
>>> fruits.remove("kiwi")
>>> fruits
['apple', 'peach', 'banana', 'pear', 'plum']
```

Note that `remove` gets rid of the **first** matching element only:

```python
>>> planets = ["Mercury", "Venus", "Venus", "Earth", "Venus"]
>>> planets.remove("Venus")
>>> planets
['Mercury', 'Venus', 'Earth', 'Venus']
```

We can also delete items by index number with the `del` command:

```python
>>> planets = ["Mercury", "Venus", "Earth", "Venus"]
>>> del planets[3]
>>> planets
['Mercury', 'Venus', 'Earth']
```

### List membership - index and in

### Finding items in a list

We can find the position of an item in a list with the `index()` method. This returns the index of the first item that matches the argument:

```python
>>> planets = ["Mercury", "Venus", "Venus", "Earth", "Venus"]
>>> planets.index("Venus")
1

>>> colours = ["red", "magenta", "blue", "cyan", "green", "yellow", "orange"]
>>> colours.index("yellow")
5

```

If the item is not in the list, the instruction will fail with a Python `ValueError`:

```python
>>> pets = ["dog", "cat", "hamster"]
>>> pets.index("gecko")
ValueError
```

### Check item in list

Just like we did with strings, we can check whether an item is in a list with the `in` operator:

```python
>>> pets = ["dog", "cat", "hamster"]
>>> "dog" in pets
True
>>> "gecko" in pets
False
```

## 🌡️ Worked Example 1 - Average Temperature 

Here is the average temperature program again. This time we are going to pass
a list of temperatures to the function.

```python

def analyseTemps(temps):
    total = 0
    index = 0
    while index < len(temps):
        total += temps[index]
    
    return total / len(temps)

# MAIN PROGRAM

average = analyseTemps([21.1, 23.4, 26.9, 24.1, 25.5, 29.0, 27.8])

print("Average temperature was:", average)

```

#### Explanation

1. `def analyse_temps(temps):` - This line defines a function named
   `analyse_temps` that takes one parameter, `temps`, which is expected to be a
   list of temperature values.

2. `total = 0` - Initializes a variable `total` to 0. This variable will be
   used to accumulate the sum of all temperatures in the list.

3. `index = 0` - Initializes a variable `index` to 0. This variable will be
   used to keep track of the current position in the list as we iterate through
   it.

4. `while index < len(temps):` - Starts a while loop that will continue as
   long as `index` is less than the length of the `temps` list. This ensures that
   we process each temperature in the list.

5. `total += temps[index]` - This line adds the current temperature to
   `total`. 

6. `index += 1` - Increments the `index` variable by 1, moving to the next
   position in the list for the next iteration of the loop.

7. `return total / len(temps)` - After the loop has processed all
   temperatures, this line calculates the average temperature by dividing the
   `total` sum by the number of temperatures in the list (`len(temps)`) and
   returns the result.

## 🗨️ Worked Example 2 - Validating a language choice

```python
def getLanguage():
    supportedLanguages = ["English", "French", "Spanish", "Arabic"]:
    looping = True
    while looping:
        language = input("Please select a language: ")
        if language in supportedLanguages
            looping = False
        else:
            print("Invalid input, try again!")
    return language
```

#### Explanation

This code defines a function `get_language()` that helps users select a language from a predefined list. Let's break it down line by line:

1. `def get_language():` - This line defines a function named `get_language` that takes no parameters.
2. `supported_languages = ["English", "French", "Spanish", "Arabic"]:` - Initializes a list called `supported_languages` containing the languages that the user can select.
3. `looping = True` - Sets a boolean variable `looping` to `True`, which will control the while loop.
4. `while looping:` - Begins a while loop that will continue to execute as long as `looping` is `True`.
5. `language = input("Please select a language: ")` - Prompts the user to input a language and stores the input in the variable `language`.
6. `if language in supported_languages` - Checks if the user's input is in the `supported_languages` list.
7. `looping = False` - If the input is valid, sets `looping` to `False`, which will exit the loop.
8. `else:` - If the input is not valid, the code proceeds to the next line.
9. `print("Invalid input, try again!")` - Informs the user that the input was invalid and prompts them to try again.
10. `return language` - Once a valid language is selected, the function returns the selected language.

````python
# q1_solution.py
'''
Return the total number of steps in the list
by adding each value using a while loop.
'''
def totalSteps(stepsList):
    total = 0
    index = 0

    while index < len(stepsList):
        total += stepsList[index]
        index += 1

    return total

# main program
steps = [4500, 6200, 7000, 8000]
print("Total steps:", totalSteps(steps))
````

````python
# q2_solution.py
'''
Repeatedly ask the user to choose a menu item.
Only accept items that appear in the menu list.
Allow the user to add multiple valid items to an order
and return the completed order list.
'''
def getValidItem(menu):
    item = input("Choose an item: ")
    while item not in menu:
        item = input("Not on the menu. Choose an item: ")
    return item

def takeOrder(menu):
    order = []
    addMore = "yes"

    while addMore == "yes":
        item = getValidItem(menu)
        order.append(item)
        addMore = input("Add another item? (yes/no): ")

    return order

# main program
menu = ["Pizza", "Pasta", "Salad", "Soup"]
finalOrder = takeOrder(menu)
print("Your order:", finalOrder)
````
