### What is an array?

We need to understand the differences between **lists** and **arrays**. 

## Python has lists, not arrays

Python lists are often wrongly called **arrays**, and they can be used as such, but they are not really arrays. Unlike Python, some programming languages like **VB.NET**, **C++**, **C#** and **Java** support genuine **arrays** as well as **lists**. 

But what's the difference?

## Arrays are fixed in size

Arrays are **static,** this means they have a **fixed size** - they cannot grow or shrink. You cannot perform **append**, **remove** or **delete** on arrays. The languages **VB.NET**, **C++**, **C#** and **Java** all provide a static array structure.

In contrast, Python lists are **dynamic** - they can grow and shrink, and support **append, remove** and **delete** operations.

> *Note:* Some languages including **C#** and **VB.NET** **also** have **dynamic** arrays, which still enforce a **single data type** but can **grow and shrink** like Python lists.

## Arrays contain only one type of data

We saw earlier that Python **lists** can contain different data types:

```python
student = ["Taha", "Hussain", 10, "Pankhurst", 104, 93.5, False] # Python list, 
# NOT REALLY AN ARRAY
```

But **arrays** may only contain data of **one type**. Here is how we define an array of seven integers called `scores` in C# :

```Csharp
int[] scores = {10, 20, 30, 40, 50, 60, 70};   // C# array of integers
```

Note that all these values have the same data type: **integer**. Of course we can define this same array as a **list** in Python:

```python
scores = [10, 20, 30, 40, 50, 60, 70] # Python list of integers,
# STILL NOT REALLY AN ARRAY!
```

### Python lists

In Python we would need to use a **list** for this data. Python lists are **dynamic**: we can append and remove elements, and we can also change the **data type** of elements or add new data of a **different data type**, like this:

```python
>>> scores[1] = "twenty"
>>> scores.append(75.5)
>>> scores
[10, 'twenty', 30, 40, 50, 60, 70, 75.5]
```

When using Python we sometimes use the term **array** when we mean a Python **list.** In the AQA exams,  a Python list is a suitable way of implementing an array, and for the rest of this Python course we will use **array** and **list** to mean the same thing.

>  

> Note: Arrays being **fixed** and enforcing a **single data type** protects data **integrity**, this means it prevents unintended changes. This is the same reason we use **constants** instead of variables where available.

## Array dimensions

So far we have met **one-dimensional** arrays, like `[10, 20, 30, 40, 50, 60, 70]`. This is a linear sequence of values, it has only **one dimension**, so we call it a **1D array**.

Later we will explore **two-dimensional** or **2D arrays**, which are needed for things like game boards, bitmaps and other information that can be arranged in a **grid** across two dimensions (e.g. horizontal and vertical).

### Pseudocode for arrays

Let's take a look at the AQA **pseudocode** for arrays, which you will see in exam papers.

# Defining 1D arrays

The pseudocode statement to allocate a one-dimensional array looks like this:

```
primes ← [2, 3, 5, 7, 11, 13]
vowels ← ['a', 'e', 'i', 'o', u']
```

## Accessing a 1D array element

Pseudocode uses the same syntax as Python to access an element:

```
>>> primes[0]
2
>>> vowels[2]
'i'
```

## Defining 2D arrays

The pseudocode statement to allocate a two-dimensional array is as follows:

```
Identifier ← [ [Exp, … ,Exp], 
                … 
               [Exp, … ,Exp] ]
```

For example:

```
ticTacToe ← [ ['.', 'O', 'O'] ,
                ['O', 'X', 'X'] ,
                ['.', 'X', '.'] ]

magicSquare ← [ [8, 1, 6] ,
                 [3, 5, 7] ,
                 [4, 9, 2] ]
```

## Accessing a 2D array element

2D Arrays need two indexes, which we can think of as row and column. We put each of them in a separate pair of square brackets, like this:

```
>>> ticTacToe[0][1]
'O'
>>> magicSquare[1][2]
7
```

This syntax is identical to Python, as we will see on the next page!

### More list operations - sum, sort

We know we can use a loop with an **accumulator pattern** to sum the contents of a list, and maybe you can write a **bubble sort** too! But Python includes some **cheat codes** so you don't have to do those things! Let's look at the `sum` and `sort` functions.

### Sum

The `sum()` function calculates the total of all numerical elements in a list. 

```python
numbers = [3, 5, 7, 9]
total = sum(numbers)
print("Sum:", total)  # Output: Sum: 24
```

### Sort

The `sort()` method sorts the elements of a list in ascending order by default. You can also sort in descending order by using the `reverse=True` parameter.

```python
>>> fruits = ["banana", "cherry", "date", "apple"]
>>> fruits.sort()
>>> print("Sorted fruits:", fruits)  
Sorted fruits: ['apple', 'banana', 'cherry', 'date']

>>> fruits.sort(reverse=True)
>>> print("Reverse sorted fruits:", fruits)  
Reverse sorted fruits: ['date', 'cherry', 'banana', 'apple']
```

> **Important:** in exams you must read the question carefully because some questions do **not allow** built-in functions like `sum() ` , or methods like `sort()`.  
>
> It is OK to use built-in functions if the question doesn't mention them, but if the question says **do not use **built-in functions, you will lose marks if you use them. 

#### About functions and methods

We call `sum()` a **function** because we call it with the list as an **argument**, like this: `sum(mylist)`. We call `sort()` a **method** because it's a special function attached to a list object, so we call it with the **dot-notation** like this: `mylist.sort()`. This distinction helps you write programs, but it isn't necessary to know these details for the exam.

### Yet more list operations - min, max

Let's look at some more useful list functions: `min` and `max`:

### Min

The `min()` function returns the smallest element in a list. This is useful for finding the minimum value in a list of numbers or the first element in alphabetical order in a list of strings.

```python
>>> numbers = [3, 5, 7, 9]
>>> min(numbers)
3
```

### Max

The `max()` function returns the largest element in a list. This is useful for finding the maximum value in a list of numbers or the last element in alphabetical order in a list of strings.

```python
grades = ['A', 'B', 'C', 'D']
>>> max(grades)
'D'
```

## 🌡️ Example 1 – highest, lowest and average temperature using min(), max() and sum()

```python
def analyseTemps(temps):
    highest = max(temps)
    lowest  = min(temps)
    total   = sum(temps)
    average = total / len(temps)

    print("Highest temperature:", highest)
    print("Lowest temperature:",  lowest)
    print("Average temperature:", average)

# MAIN PROGRAM
week = [21.1, 23.4, 26.9, 24.1, 25.5, 29.0, 27.8]
analyseTemps(week)
```

### Explanation

* `max()` finds the highest temperature.

* `min()` finds the lowest temperature.

* `sum()` adds all temperatures together.

* The average is calculated using the total divided by the number of values.

Create two new lists:
one containing all the even numbers
and one containing all the odd numbers.

nums = [3, 10, 7, 4, 12, 9]

evens = []
odds = []

i = 0
while i < len(nums):
    if nums[i] % 2 == 0:
        evens.append(nums[i])
    else:
        odds.append(nums[i])
    i += 1

print("Evens:", evens)

## 🚆 Question 3 – How Far Can I Travel on a Railway Line?

```python
'''
Write a function that takes one parameter, 
the number of minutes available for travel, and returns
the index of the last station you can reach
'''
times = [3, 12, 8, 21, 11, 15]

def maxStation(minutes):
    total = 0
    station = 0

    i = 0
    while i < len(times):
        if total + times[i] <= minutes:
            total += times[i]
            station += 1
        else:
            break
        i += 1

    return station

# Example
print(maxStation(18))   # Expected: 2
```
