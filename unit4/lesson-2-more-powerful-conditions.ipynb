### Using AND and OR

Remember the party decision program?

```python
if roomTidy == True and homeworkDone == True:
    print("go to party")
```

The first thing to note is that `== True` is not needed in this code. That's because if an operand is already **Boolean**, we can use it directly in our expression. If the operand is already **True**, then the condition becomes `True == True`, which is also **True**, and if it's **False**, then `False == True` is also **False**! Let's simplify the code:

```python
if roomTidy and homeworkDone:
    print("go to party")
```

Remember this later when you are writing logical expressions with Boolean variables or Boolean expressions. 

But back to the party, what if Dad relents and says "OK, you can go if you **either** tidy your room **or** do your homework today". That looks like this:

```python
if roomTidy or homeworkDone:
    print("go to party")
```

And let's change the pizza question to "Dan will buy pizza if he finishes work before 11 and either has more than five pounds in his pocket, **or** has remembered his debit card":

```python
if finishedWork < 11 and (cashInPocket > 5.0 or debitCardInPocket)
    print("buy pizza")
```

We have used brackets above to ensure that Python evaluates the **or** operation first, then the **and** operation. This is important because the usual **order of operations **would perform **and** before **or,** but brackets are always evaluated first, just as in arithmetic operations (BIDMAS).

### Using AND, OR and NOT

The final Boolean operator you need to know is **not**. Unlike **and** and **or**, the **not** operator takes only one **operand**, and **inverts** it, changing `True` to `False` or `False` to `True`. 

### Using NOT

Consider our party question again. This time Mum says "you can go to the party if you have **not** been given any detentions this term":

```python
if not hadDetentions:
    print("go to party")
```

In the above code, `hadDetentions` is a Boolean variable. But we can use any **Boolean expression** in our conditions, including those with arithmetic comparison operators like `>` and `!=`. If we have an integer variable **`numberOfDetentions`** then the code might look like this:

```python
if numberOfDetentions == 0:
    print("go to party")
```

### Mixing AND and NOT

Dad has a word with Mum and relaxes the rule to "no more than two detentions, as long as you also wash the car, and don't argue with your brother". This would become:

```python
if numberOfDetentions == 0 and carWashed and not arguedWithBrother:
    print("go to party")
```

Notice that either side of every **and** statement, and following the **not** statement is a complete **Boolean expression**, either a Boolean variable (**`carWashed`**) or a comparison (**`numberOfDetentions == 0`**).

## Writing a Boolean function

Let's look at one last example, and this time let's write a function that returns a Boolean value. 

A ticket barrier will open when you tap your off-peak travelcard, as long as all of these are true:

* the card is valid

* the balance on the card is at least £1.00 

* the time of day is not before 9am. 

Write a function that returns `True` if the barrier should open, and `False` if not.

```python
def checkSmartcard(cardValid, cardBalance, timeOfDay):
  if cardValid and cardBalance >= 1.00 and not (timeOfDay <= 9)
    return True
  else:
    return False
```

Notice there are sometimes several ways of writing conditions, we could write `timeOfDay > 9` instead, but `not (timeOfDay <= 9)` reflects the instructions given in the question more closely.

## Returning a Boolean directly

In the above example, we don't actually need an if statement! The Boolean expression evaluates to True or False already, so we can just return that directly and achieve the same result:

```python
def checkSmartcard(cardValid, cardBalance, timeOfDay):
  return cardValid and cardBalance >= 1.00 and not (timeOfDay <= 9)
```

The expression after `return` is evaluated, and the resulting value is returned from the function, which is exactly the same value as in the original `if-else` version!

### Operator precedence

Just like in maths, Python has rules about which parts of a logical expression get evaluated first. These rules are called *operator precedence*. Like BIDMAS, Brackets are always evaluated first, then AND, OR and NOT, in that order. If you have arithmetic, comparisons **and** Boolean operations, then the full operator precedence rule is:

| **Operator (high to low)** | **Examples**                        |
| -------------------------- | ----------------------------------- |
| **Brackets / Parentheses** | `(x + y)`, `(a > b and c)`          |
| **Arithmetic operators**   | `**`, `*`, `/`, `//`, `%`, `+`, `-` |
| **Comparison Operators**   | `<`, `<=`, `>`, `>=`, `==`, `!=`    |
| **NOT**                    | `not x`                             |
| **AND**                    | `x and y`                           |
| **OR**                     | `x or y`                            |

You could use the acronym BACNAO or "Boole ate cheese, never ate olives"!

### 🔢 Worked Example 1: Number checker

Let's create a function that checks if a number is a **two-digit number that is divisible by either two or three**, for example 10, 12, 14, 15, 16, 18, 20, 21 .. 96, 98, 99. We'll use logical operators `and`, `or`, and `not` to achieve this. 

Firstly let's work out some conditions we can use. 

* **Having two digits** just means the number is between 10 and 99 inclusive. 
* **Divisible by two** means that when we divide by two and take the **remainder**, this is **zero**. 
* Likewise, **divisible by three** means when we divide by three the **remainder** is **zero**. 
* We can use **MOD** or the Python operator **`%`** for the 'divisible by' checks

```python
def checkNumber(num):
    # Check if the number is between 10 and 99, and divisible by 2 or 3
    if (num >= 10 and num <= 99) and (num % 2 == 0 or num % 3 == 0):
        return "Yes"
    else:
        return "No"

# main program  test the function
print(checkNumber(12))  # Output: Yes
print(checkNumber(15))  # Output: Yes
print(checkNumber(25))  # Output: No
print(checkNumber(99))  # Output: Yes
```
