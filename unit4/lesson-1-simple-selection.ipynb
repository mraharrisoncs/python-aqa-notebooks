### Why Selection?

## Sequences

Programs that use only sequences of operations are not very useful. Remember this program?

```python
colour = input("What is your favourite colour?")
print("I like", colour, "too!")
```

But what if the computer's favourite colour is blue, and the user enters "red"? In this case we might want the program to say "Good choice, but I like blue best".

## Selection

Here's how we do that:

```python
colour = input("What is your favourite colour?")
if colour == "blue":
  print("I like", colour, "too!")
else:
  print("Good choice, but I like blue best")
```

Python gives us the keywords `if` and `else` to control the **flow** of the program. In the above example, if the user enters "blue", the program will output "I like blue too!", if they enter anything else it will say "Good choice..."

### Structure of IF-ELSE

<br />

Remember this program? Let's look at this code in more detail. 

```python
weather = input("What is the weather today? ")
if weather == "raining":
    print("Take an umbrella")
else:
    print("Good to go!")
```

### Explanation

The second line tells Python we need to evaluate the **expression** `weather == "raining"` to decide which line to execute next  

> Tip: remember an **expression** is any valid sequence of **terms** and **operators** that can be **evaluated** to obtain a **result**. We met **arithmetic expressions** earlier in the course, which always evaluate to numbers. Here we have a **Boolean expression**, which evaluates to either **True** or **False.**

* The expression, `weather == "raining"` will evaluate to `True` if the variable `weather` holds the string "raining". 

* If the expression is `True`, Python executes the next line. 

* If the expression is `False`, Python skips all the indented code, and because it finds an `else:` keyword it executes the code block after this. (`else:` is optional, as we will see later.)

> *Exam tip:* An if-else statement is called a **binary** conditional statement, because there are **two** branches of code to choose from, and binary means ***two-ness*****.** 

## '=' versus '==' 

In `if` statements, we always use two equals signs, not just one. This is because `==` means something different to `=` .

In Unit 1 we learned about variables and how to give them values. You may recall **assignment** means giving a value to a variable, like this:

```python
name = "Ada"       # this is an assignment statement, so a single equals sign is needed
```

But now we are performing **selection**, which uses **Boolean expressions**, where we compare two values to see if they are equal. For this we need `==` which means **is equal to**. 

```python
if name == "Ada":  # this is a Boolean expression, so we need to use two equals signs
```

### Boolean Expressions - comparing things

### Boolean Expressions in IF statements

Remember, a Boolean expression is an expression that evaluates to either `True` or `False`. We use the result to control the flow of a program by using selection statements like `if` and `else`. We were looking at this example program:

```python
weather = input("What is the weather today? ")
if weather == "raining":
  print("Take an umbrella")
else:
  print("Good to go!")
```

The Boolean expression here is `weather == "raining"`, which always evaluates to either True or False, nothing else. We can use **any** Boolean expression as the condition on an `if` statement. Anything that evaluates to `True` or `False` can control an `if` statement.

#### Comparison Operators

In the above example, `==` is a **comparison operator** because it compares two values, in this case the contents of the variable `weather` and the string literal "raining". Putting two **operands** like this either side of a comparison operator makes a valid **Boolean expression,** which evaluates to a **Boolean value**: either `True` or `False`. 

Here are the common comparison operators in Python:

* `==`: Equal to

* `!=`: Not equal to

* `>`: Greater than

* `>=`: Greater than or equal to

* `<`: Less than

* `<=`: Less than or equal to

#### Writing a Boolean Expression

A Boolean expression involves a comparison **operator** and two **operands,** one either side. Read through these examples and consider why we get the output shown:

```python
>>> x = 10
>>> y = 20

>>> print(x == y)
False

>>> print(x != y)
True

>>> print(x > y)
False

>>> print(x <= y)
True
```

### Programming if-else

A **Boolean expression** is used in an `if-else` statement to decide which block of code should be executed. 

## Can you vote?

Here's another example that uses `>=` (greater than or equal to), and this time we have wrapped the code in a **function** definition with `def`:

```python
def canVote(age):
    if age >= 18:
        print("You are eligible to vote")
    else:
        print("You are not eligible to vote")

# main program
myAge = int(input("Age? "))
print(canVote(myAge))
```

### Explanation

In this example, the Boolean expression `age >= 18` is evaluated. If `True`, the program prints "You are eligible to vote". Otherwise, it prints "You are not eligible to vote".

## Speed warning function

 

Here is another example, a **function** that takes speed as a parameter, and returns a string. I've also used a variable as if it is a **constant** at the top of the code, remembering to use all capitals. This 'constant' allows us to easily change the value later, for example if the speed limit becomes 20.

```python
SPEED_LIMIT = 30      

def speedCheck(currentSpeed):
    if currentSpeed <= SPEED_LIMIT:
        return "Safe driving, keep going."
    else:
        return "Too fast, slow down!"

# main program
speed = int(input("Current speed? "))
print(speedCheck(speed))
```

### Explanation

This time, the Boolean expression has two **variables** as the operands, and `less than or equal to` as the comparison operator. The expression `currentSpeed <= SPEED_LIMIT` is evaluated. If `True`, the **function** will return the string "Safe driving, keep going." Otherwise, it returns "Too fast, slow down!"

Back in the main program, the return value is printed, because we wrote `print(speedCheck(speed)).`

## Pseudocode notes

Programming languages use combinations of basic symbols such as `!=` and `<=` because these symbols are available in ASCII and can be entered directly on computer keyboards. In AQA exams, however, we use the standard mathematical symbols for comparison operators: `=`, `≥` and `≤`. Also we use the keywords `IF`, `THEN`, `ELSE` and `ENDIF`, and don't forget we use `OUTPUT` for `print()`, so the 'weather' example looks like this...

```python
IF weather = "raining" THEN
    OUTPUT "Take an umbrella"
ELSE
    OUTPUT "Good to go!"
ENDIF
```

...the voting example like this...

```python
IF age ≥ 18 THEN
    OUTPUT "You are eligible to vote"
ELSE
    OUTPUT "You are not eligible to vote"
ENDIF
```

...and the speed check example like this:

```python
IF currentSpeed ≤ SPEED_LIMIT THEN
    RETURN "Safe driving, keep going."
ELSE
    RETURN "Too fast, slow down!"
ENDIF
```

### IF without ELSE

# IF without ELSE

We have talked about `if-else` so far, because we usually write selection with two branches. The Boolean expression decides which branch is executed, like this:

```python
weather = input("What is the weather today? ")
if weather == "raining":
    print("Take an umbrella")
else:
    print("Good to go!")
```

Which looks like this in a flowchart:

1.00

But sometimes we do not need the second branch. Here is an example:

Write a program to ask if someone wants milk in their tea. Output instructions to make the drink:

```python
print("make the tea")
milk = input("Do you want milk?")
if milk == "yes":
    print("add milk")
print("stir and drink the tea")
```

Notice that there is no `else` statement, because there is no alternative path the program needs to take. 

Here is another example.  Train tickets are 30% cheaper with a student railcard (or 70% of the standard price!)

```python
fare = 10.00
railcard = input("Do you have a railcard")
if railcard == "yes":
    fare = fare * 0.7
print("fare =", fare)
```

Again there is no need for an `else` statement, because the fare remains at £10 without a railcard, there is no need to do anything to it.

> *Exam tip:* this is called a **unary** conditional, because it only has **one** branch, and unary means ***one-ness***.
