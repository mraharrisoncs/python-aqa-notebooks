### Nested ifs

### Introduction to Nested IF Statements

Remember the party decision program?

```python
if roomTidy and homeworkDone:
    print("go to party")
```

We could have written this as two separate if statements, one inside the other. This is called **nesting** if statements.

```python
if roomTidy:
    if homeworkDone:
        print("go to party")
```

When the above program runs, Python first checks the value of `room_tidy`, and if this is True, it executes the second `if` statement. This then checks the Boolean `homework_done`. If this is also True, it prints "go to party". The behaviour is identical to the first version that used `and` to make a compound condition.

In Python, nested if statements allow you to check multiple conditions within another condition. This is useful when you need to make decisions based on more than one criterion. A nested **if** is simply an **if** statement inside another **if** statement.

### Syntax of Nested IF

Here's the general structure of a nested if statement:

```python
if condition1:
    # Executes when condition1 is true
    if condition2:
        # Executes when condition2 is true
        print("Both conditions are true")
    else:
        # Executes when condition2 is false
        print("Condition1 is true, but condition2 is false")
else:
    # Executes when condition1 is false
    print("Condition1 is false")
```

### Example 1: test scores

Let's write a nested if statement to determine if a student has passed a course based on their test score and attendance. A passing score is 50, and the student must have attended at least 75% of classes.

```python
if score >= 50:
    if attendance >= 75:
        print("You passed the course!")
    else:
        print("Your test score was good but your attendance was too low")
else:
    print("Your test score was too low")
```

### Indenting

Remember that Python uses indenting to show where a code block starts and ends. In the above example, the inner **if** statement is indented by four spaces to show that it is inside the outer if statement. Within this indented block you **must** continue to indent the code correctly, hence the inner **if** statement branches are indented just as they would be if they were not nested within the outer statement.

### Compound conditions vs. nested ifs

We saw how **`and`** and **`or`** allow us to join conditions together on a single **`if`** statement. So why would you ever need nested ifs? Because they give us more control. We can have multiple branches of code and execute each one conditionally.

## How would you like your coffee?

A vending machine offers many choices of drink. But not all choices apply to all drinks. This is a perfect candidate for nested if statements. Let's take a look.

The vending machine algorithm looks like this:

1. What drink, Tea, Coffee or Hot Chocolate?
2. If Tea or Coffee chosen:

   a. Milk (y/n)?
   \
   b. Sugar (y/n)?

       i. If sugar chosen, how many sugars (1-3)?
3. Else (Hot Chocolate):

   a. Chocolate sprinkles (y/n)?
4. Dispense drink

Note that the conditions under step 2 apply only to a choice of tea or coffee, you can't put extra milk or sugar in hot chocolate. Likewise you can't ask for sprinkles on tea or coffee (sorry!).

We can write this in AQA-style pseudocode for more clarity:

```python
OUTPUT "What drink, Tea, Coffee or Hot Chocolate?"
INPUT drink

IF drink = "Tea" OR drink = "Coffee" THEN
    OUTPUT "Milk (y/n)?"
    INPUT milkOption
    OUTPUT "Sugar (y/n)?"
    INPUT sugarOption
    IF sugarOption = "y" THEN
        OUTPUT "How many sugars (1-3)?"
        INPUT numberOfSugars
    ELSE
        numberOfSugars = 0
ELSE
    OUTPUT "Chocolate sprinkles or no sprinkles?"
    INPUT sprinklesOption
ENDIF

OUTPUT "Dispense drink"
```

Let's write this in Python now:

```python
# Ask for drink choice
drink = input("What drink, Tea, Coffee or Hot Chocolate? ")

    if sugarOption == "y":
        numberOfSugars = int(input("How many sugars (1-3)? "))
    else:
        numberOfSugars = 0
else:
    sprinklesOption = input("Chocolate sprinkles or no sprinkles? ")

print("Dispense drink")
```

### Introducing elif

### Understanding if-elif-else

In Python, the `if-elif-else` statement is a powerful tool for decision-making. It allows you to check multiple conditions and execute different blocks of code based on which condition is true. This is particularly useful when you have more than two possible outcomes. Consider this function that returns a grade based on the score passed in as a parameter. 

```python
def grade(score):
  if score >= 80:
    return 'A'
  else:
    if score >= 60
      return 'B'
    else:
      if score >= 40
        return 'C'
      else:
        return 'F'
```

After the first conditional, we have nested each if statement in the **else** branch of the previous one. This works, but it's really messy, and the indentation is out of control! Fortunately, Python gives us an easier way to do this. The **elif** statement is short for **else if**, and the above function can be simplified like this:

```python
def grade(score):
  if score >= 80:
    return 'A'
  elif score >= 60
    return 'B'
  elif score >= 40
    return 'C'
  else:
    return 'F'
```

### Syntax of if-elif-else

The general syntax of an `if-elif-else` statement is as follows:

```python
if condition1:
    # Code to execute if condition1 is true
elif condition2:
    # Code to execute if condition1 is false and condition2 is true
elif condition3:
    # Code to execute if condition1 and condition2 are false and condition3 is true
'''
add more elif branches as necessary here
'''
else:
    # Code to execute if none of the above conditions are true
```

* **`if`**: The initial condition to check. If this condition is true, the corresponding block of code is executed.

* **`elif`**: Short for "else if," this allows you to check additional conditions if the previous ones are false.

* **`else`**: The block of code that runs if none of the preceding conditions are true.

## 📅 Example 1: Date-Related Decision

Let's create a simple program that inputs the day of the week and prints a message based on whether it's a weekday or the weekend.

```python
# Get the current day of the week
currentDay = input("What day is it? (Mon, Tue, Wed...)? ")

if currentDay in ["Mon", "Tue", "Wed", "Thu", "Fri"]:
    print("It's a weekday. Time to work or study!")
elif currentDay == "Sat":
    print("It's Saturday. Enjoy your weekend!")
else:
    print("It's Sunday! Relax and recharge for the week ahead.")
```

#### Explanation:

* We input the current day of the week.

* The `if` statement checks if the day is a weekday.

* The `elif` statement checks if it's "Sat" meaning Saturday.

* The `else` statement handles the case for Sunday.
